```python

import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

```

```python
groot_abstraction = pd.read_csv("Groot_abstraction_TDS.csv", parse_dates= ['Date'])
f = plt.figure()
f.set_figwidth(18)
f.set_figheight(7)
ax1 = sns.lineplot(data=groot_abstraction, x='Date', y = "Simulated TDS", label = "", color = "red")
ax1 = sns.lineplot(data=groot_abstraction, x='Date', y = "Threshold", label = "", color = "purple")
ax1.tick_params(axis='x', labelsize=20)
ax1.tick_params(axis='y', labelsize=20)
ax2 = ax1.twinx() 
ax2 = sns.lineplot(x = "Date", y = "Output", color = "blue", data=groot_abstraction, label = "")               
start_date = datetime.strptime('01/01/1920 00:00','%d/%m/%Y %H:%M')
end_date = datetime.strptime('01/01/2010 00:00','%d/%m/%Y %H:%M')

plt.xlim([start_date, end_date])
plt.tick_params(axis='both', which='major', labelsize=15)
ax2.tick_params(axis='x', labelsize=20)
ax2.tick_params(axis='y', labelsize=20)
#plt.ylim([0, 0.15])
#plt.xlabel('Date', fontsize=20)
#plt.ylabel('Total dissolved solids (mg/l)', fontsize=20)
#ax1.legend(loc='lower left', fontsize = 20)
#ax2.legend(loc='lower right', fontsize = 20)
#plt.tick_params(axis='both', which='major', labelsize=15)
plt.savefig('Grootdraai_abstraction_TDS.png', dpi = 300)
plt.show()
```
